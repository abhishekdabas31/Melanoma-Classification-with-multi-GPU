```python
## Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Time

```python
dict_time = {"CPU": 5279.6492 ,
         "GPU" : 1776.9850}

plt.plot(*zip(*sorted(dict_time.items())), '--bo')
plt.title("Time VS System")
plt.xlabel("Epochs")
plt.ylabel("Time(sec)")
plt.show()
```

## Accuracy

```python
dict_acc = {"CPU": 27.96 ,
         "GPU" : 39.11}

plt.plot(*zip(*sorted(dict_acc.items())), '--bo')
plt.title("Accuracy VS system")
plt.xlabel("System")
plt.ylabel("Accuracy(%)")
plt.show()
```

# Loss

```python
dict_loss = {"CPU":  0.3849,
         "GPU" : 0.1154}

plt.plot(*zip(*sorted(dict_loss.items())), '--bo')
plt.title("Loss VS System")
plt.xlabel("System")
plt.ylabel("Loss")
plt.show()
```
