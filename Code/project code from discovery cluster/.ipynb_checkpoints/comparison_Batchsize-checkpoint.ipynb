```python
## imports
import matplotlib.pyplot as plt
import pandas as pd
```

# Time comparison

```python
dict_time = {"128" :687.631,
            "256": 1262.101 ,
             "512" : 1360.3255}

plt.plot(*zip(*sorted(dict_time.items())), '--bo')
plt.title("Time VS Batch_size")
plt.xlabel("Batchsize")
plt.ylabel("Time(sec)")
plt.show()
```

## Loss comparison

```python
dict_loss = {"128":0.079,
            "256": 0.085 ,
             "512" : 0.085}

plt.plot(*zip(*sorted(dict_loss.items())), '--bo')
plt.title("Loss VS Batch_size")
plt.xlabel("Batchsize")
plt.ylabel("Loss")
plt.show()
```

## accuracy comparison

```python
dict_acc = { "128" : 82.85,
            "256": 78.003 ,
             "512" : 66.3}


plt.plot(*zip(*sorted(dict_acc.items())), '--bo')
plt.title("Accuracy VS Batch_size")
plt.xlabel("Batchsize")
plt.ylabel("Accuracy")
plt.show()
```
